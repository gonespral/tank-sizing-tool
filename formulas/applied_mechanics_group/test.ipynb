```python
from discontinity import *
from math import *
import seaborn as sns
import pandas as pd

# Initiate test values

p =45E5
r = 0.24
R = r + 0.00000000001
t_0 = 0.004
t_1 = 0.004
rho_1 = 0.33
```

```python
stresses = []

for psi in range(0, 90, 1):
    psi = radians(psi)

    s_hoop, s_meridional, s_max  = discontinuity_forces(p=p,
                                  r=r,
                                  R=R,
                                  t_0=t_0,
                                  t_1=t_1,
                                  rho_1=rho_1,
                                  psi=psi)

    stresses.append([psi, s_hoop, s_meridional, s_max])

df = pd.DataFrame(stresses, columns=['psi', 's_hoop', 's_meridional', 's_max'])

df
```
