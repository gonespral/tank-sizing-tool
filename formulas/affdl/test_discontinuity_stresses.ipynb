```python
import discontinuity_stresses as ds
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
```

```python
df = pd.DataFrame(columns=['hoop_normal_stress', 'meridional_shear_stress', 'maximum_meridional_bending_stress', 'maximum_hoop_bending_stress', 'Q_0', 'M_0', 'w_m1', 'w_m2', 'resultant'])

for x in np.linspace(0, 1, 1000):
    res = ds.find_stress_state(pressure=45E5,
                               phi_0=np.deg2rad(90),
                               thickness_dome=0.004,
                               thickness_cylinder=0.004,
                               poisson_ratio=0.33,
                               youngs_modulus=69E9,
                               radius=0.24,
                               x=x,
                               dome_type="figure_of_revolution",
                               find_in="dome",
                               R_t=0.24,
                               R_m=0.24)
    res["x"] = x

    df = df.append({str(key): res[key] for key in res}, ignore_index=True)

```

```python
df
```

```python
sns.set_theme(style="white")

sns.lineplot(data=df,
            x="x",
            y="hoop_normal_stress",
            label="hoop_normal_stress")

plt.show()

sns.lineplot(data=df,
            x="x",
            y="meridional_shear_stress",
            label="meridional_shear_stress")

plt.show()

sns.lineplot(data=df,
            x="x",
            y="maximum_meridional_bending_stress",
            label="maximum_meridional_bending_stress")

plt.show()

sns.lineplot(data=df,
            x="x",
            y="maximum_hoop_bending_stress",
            label="maximum_hoop_bending_stress")

plt.show()

```

```python
sns.lineplot(data=df,
            x="x",
            y="resultant",
            label="Resultant stress")

plt.ylabel("Stress (Pa)")
plt.xlabel("x (m)")

plt.show()
```
